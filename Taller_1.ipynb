Dado que tenemos más de 50mil datos, vamos a borrar todas las observaciones con menos de 100 nulos

```python
#sacamos las columnas del df
for column in df.columns:
    #Sacamos cuantos nulos tiene
    num_nan_values = df[column].isna().sum()
    #Si tiene menos de 100 pero más de 0 nulos, le hacemos drop a esas filas
    if num_nan_values < 100 and num_nan_values >0:
        df = df.dropna(subset=[column])
```

Tenemos dos columnas de etnia, podemos ver que tenemos si y no, y a cual pertenecen. En la segunda columna tenemos varias columnas vacias, pero estas probablemente representan las observaciones que no tienen etnia, por ahora los marcaremos todos con un mismo valor para más tarde hacer les label o one hot encoding

```python
from pandas.core.arrays.sparse.array import notna
for index, row in df.iterrows():
    # En el caso que no tenga etnia y tenga nan en etnia, pondremos 0 en la columna de etnia
    if row['ESTU_TIENE_ETNIA']=='NO' and pd.isna(row['ESTU_ETNIA']):
        df.loc[index, 'ESTU_ETNIA'] = '0'
    #En el caso que tenga nan en tiene etnia, pero tenga etnia, pondremos 1
    if pd.isna(row['ESTU_TIENE_ETNIA']) and pd.notna(row['ESTU_ETNIA']):
        df.loc[index, 'ESTU_TIENE_ETNIA'] = 'SI'
    #En el caso de que ambos sean nulos, pondremos que 0 en etnia y no en tiene etnia
    #if pd.isna(row['ESTU_TIENE_ETNIA']) and pd.isna(row['ESTU_ETNIA']):
      #df.loc[index, 'ESTU_ETNIA'] = '0'
      #df.loc[index, 'ESTU_TIENE_ETNIA'] = 'NO'
    #En el caso de que si tiene etnia, pero no etnia es nan, tambien pondremos 1
    if pd.notna(row['ESTU_TIENE_ETNIA']) and pd.isna(row['ESTU_ETNIA']):
      df.loc[index, 'ESTU_ETNIA'] = '1'
```

```python
df.isna().sum()
```

Nos encontramos con un pequeño problema, para 4313 personas no sabemos si tienen o no etnia, ni cual etnia tienen. Dado que es una buena cantidad de datos, usaremos una tecnica de imputación

```python
df['ESTU_TIENE_ETNIA'].mode()
```

Dado que la moda es no, usaremos este valor para reemplazar los 4313 datos faltanes

```python
from pandas.core.arrays.sparse.array import notna
for index, row in df.iterrows():
    #En el caso de que ambos sean nulos, pondremos que 0 en etnia y no en tiene etnia
    if pd.isna(row['ESTU_TIENE_ETNIA']) and pd.isna(row['ESTU_ETNIA']):
      df.loc[index, 'ESTU_ETNIA'] = '0'
      df.loc[index, 'ESTU_TIENE_ETNIA'] = 'NO'
```

Parece ser que estas columnas son 1 hot encoding, así que vamos a asumir que los nan = 0 y x = 1 (Esto se puede ver mejor si miramos el dataset en excel)

```python
for index, row in df.iterrows():
    #Si el dato es nan, lo marcamos como 0, sino, lo marcamos como 1
    if pd.isna(row['ESTU_LIMITA_MOTRIZ']):
        df.loc[index, 'ESTU_LIMITA_MOTRIZ'] = 0
    else:
       df.loc[index, 'ESTU_LIMITA_MOTRIZ'] = 1
    if pd.isna(row['ESTU_LIMITA_INVIDENTE']):
        df.loc[index, 'ESTU_LIMITA_INVIDENTE'] = 0
    else:
       df.loc[index, 'ESTU_LIMITA_INVIDENTE'] = 1
    if pd.isna(row['ESTU_LIMITA_CONDICIONESPECIAL']):
        df.loc[index, 'ESTU_LIMITA_CONDICIONESPECIAL'] = 0
    else:
       df.loc[index, 'ESTU_LIMITA_CONDICIONESPECIAL'] = 1
    if pd.isna(row['ESTU_LIMITA_SORDO']):
        df.loc[index, 'ESTU_LIMITA_SORDO'] = 0
    else:
       df.loc[index, 'ESTU_LIMITA_SORDO'] = 1
    if pd.isna(row['ESTU_LIMITA_AUTISMO']):
        df.loc[index, 'ESTU_LIMITA_AUTISMO'] = 0
    else:
       df.loc[index, 'ESTU_LIMITA_AUTISMO'] = 1

```

Dado que hay datos repetidos, no podemos simplemente poner todos como 1, dado que el colegio de donde salio el estudiante podria tener algun efecto en su resultado.

Lo que haremos es ver si en esta columna hay un nulo, pero en la siguiente columna hay un dato, le daremos un valor numerico para hacer label encoding más tarde

```python
for index, row in df.iterrows():
    #Si el dato es nan, lo marcamos como 0 en termino colegio si termino el colegio en otro colegio
    if pd.isna(row['ESTU_COLE_TERMINO']) and pd.notna(row['ESTU_OTROCOLE_TERMINO']):
        df.loc[index, 'ESTU_COLE_TERMINO'] = 0
    #Si termino en el primer colegio y el otro colegio es nan, pondremos nan
    if pd.isna(row['ESTU_OTROCOLE_TERMINO']) and pd.notna(row['ESTU_COLE_TERMINO']):
        df.loc[index, 'ESTU_OTROCOLE_TERMINO'] = 0
```

Todos estos datos son nulos, asi que se asume que el estudiante no termino el colegio, sinembargo, estos datos aun nos pueden dar bastante información, asi que les asignaremos otro numero (1) para luego hacer 1 hot encoding

```python
for index, row in df.iterrows():
    if pd.isna(row['ESTU_COLE_TERMINO']) and pd.isna(row['ESTU_OTROCOLE_TERMINO']):
       df.loc[index, 'ESTU_COLE_TERMINO'] = 1
       df.loc[index, 'ESTU_OTROCOLE_TERMINO'] = 1
```
